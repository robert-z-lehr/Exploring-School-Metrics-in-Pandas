# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd

# Files to Load
str_school_data_to_load = "Resources/schools_complete.csv"
str_student_data_to_load = "Resources/students_complete.csv"

# Read the respective school and student data files and store them into Pandas DataFrames
df_school_data = pd.read_csv(str_school_data_to_load)
df_student_data = pd.read_csv(str_student_data_to_load)

# Combine the data into a single dataset 
df_school_data_complete = pd.merge(df_student_data, df_school_data, how="left", on=["school_name", "school_name"])
df_school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
array_school_count = df_school_data_complete["school_name"].unique()
int_school_count = len(array_school_count)
print(f"There are {int_school_count} unique schools.")
```

```python
# Calculate the total number of students
int_student_count = df_school_data_complete["student_name"].count()
print(f"There are {int_student_count:,} students.")
```

```python
# Calculate the total budget
int_total_budget = df_school_data["budget"].sum()
print(f"The total budget for the district and charter schools combined is ${int_total_budget:,}!")
```

```python
# Calculate the average (mean) math score
float_avg_math_score = df_school_data_complete["math_score"].mean()
print(f"The average math score is {float_avg_math_score:.2f}!")
```

```python
# Calculate the average (mean) reading score
float_avg_reading_score = df_school_data_complete["reading_score"].mean()
print(f"The average reading score is {float_avg_reading_score:.2f}!")
```

```python
# Calculate the percentage of students who passed math (values found in the df_school_data_complete["math_score"] column >= 70)
int_passing_math_count = df_school_data_complete[(df_school_data_complete["math_score"] >= 70)].count()["student_name"]
float_passing_math_percentage = float(int_passing_math_count) / float(int_student_count) * 100
print(f"{float_passing_math_percentage:.2f}% of students passed their math classes.")
```

```python
# Calculate the percentage of students who passed reading (values found in the df_school_data_complete["reading_score"] column >= 70) 
int_passing_reading_count = df_school_data_complete[(df_school_data_complete["reading_score"] >= 70)].count()["student_name"]
float_passing_reading_percentage = float(int_passing_reading_count) / float(int_student_count) * 100
print(f"{float_passing_reading_percentage:.2f}% of students passed their reading classes.")
```

```python
# Calculate the percentage of students that passed both math and reading
int_passing_math_reading_count = df_school_data_complete[(df_school_data_complete["math_score"] >= 70) & (df_school_data_complete["reading_score"] >= 70)].count()["student_name"]
series_overall_passing_rate = int_passing_math_reading_count /  float(int_student_count) * 100
series_overall_passing_rate
print(f"{series_overall_passing_rate:.2f}% of students passed both their reading and math classes.")
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({'Total Schools': [int_school_count],
                                 'Total Students': [f'{int_student_count:,}'],
                                 'Total Budget': [f'${int_total_budget:,.2f}'],
                                 'Average Math Score': [f'{float_avg_math_score:.2f}'],
                                 'Average Reading Score': [f'{float_avg_reading_score:.2f}'],
                                 '% Passing Math': [f'{float_passing_math_percentage:.2f}%'],
                                 '% Passing Reading': [f'{float_passing_reading_percentage:.2f}%'],
                                 '% Overall Passing': [f'{series_overall_passing_rate:.2f}%']})

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select the school type
series_school_types = df_school_data.set_index(["school_name"])["type"]
print(series_school_types.to_string())
```

```python
# Calculate the total student count
series_per_school_counts = df_student_data["school_name"].value_counts()
print(series_per_school_counts.to_string())
```

```python
# Calculate the total school budget
series_per_school_budget = df_school_data_complete.groupby(["school_name"])["budget"].mean()

# Claculate the per capita spending
series_per_school_capita = series_per_school_budget / series_per_school_counts

#Concatenate series horizontally to display
df_horizontal_display_budget_capita = pd.concat([series_per_school_budget,series_per_school_capita], axis=1, keys = ["Budget", "Capita"])
df_horizontal_display_budget_capita
```

```python
# Calculate the average test scores
series_per_school_math = df_school_data_complete.groupby(["school_name"])["math_score"].mean()
series_per_school_reading = df_school_data_complete.groupby(["school_name"])["reading_score"].mean()

#Concatenate series horizontally to display
df_horizontal_display_average_scores = pd.concat([series_per_school_math,series_per_school_reading], axis=1, keys = ["Avg Math Scores", "Avg Reading Scores"])
df_horizontal_display_average_scores
```

```python
# Calculate the number of schools with math scores of 70 or higher
df_school_passing_math = df_school_data_complete[(df_school_data_complete['math_score'] >= 70)]

# Prints each unique school name, which satisfy this condition
print(df_school_passing_math['school_name'].unique())

# Prints the length of the dataframe
len(df_school_passing_math['school_name'].unique())
```

```python
# Calculate the number of schools with reading scores of 70 or higher
df_school_passing_reading = df_school_data_complete[(df_school_data_complete['reading_score'] >= 70)]

# Prints each unique school name, which satisfy this condition
print(df_school_passing_reading['school_name'].unique())

# Prints the length of the dataframe
len(df_school_passing_reading['school_name'].unique())
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
df_passing_math_and_reading = df_school_data_complete[
    (df_school_data_complete["reading_score"] >= 70) & 
    (df_school_data_complete["math_score"] >= 70)]

# Prints each unique school name, which satisfy this condition
print(df_passing_math_and_reading['school_name'].unique())

# Prints the length of the dataframe
len(df_passing_math_and_reading['school_name'].unique())
```

```python
# Use the provided code to calculate the passing rates
series_per_school_passing_math = df_school_passing_math.groupby(["school_name"]).count()["student_name"] / series_per_school_counts * 100
series_per_school_passing_reading = df_school_passing_reading.groupby(["school_name"]).count()["student_name"] / series_per_school_counts * 100
series_overall_passing_rate = df_passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / series_per_school_counts * 100

#Concatenate series horizontally to display
df_horizontal_display_math_reading_both = pd.concat(
    [series_per_school_passing_math, series_per_school_passing_reading, series_overall_passing_rate], 
    axis=1, keys = ["Passing Math", "Passing Reading", "Passing Both"])
df_horizontal_display_math_reading_both
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
dict_school_summary = {
    'School Type': series_school_types,
    'Total Students': series_per_school_counts,
    'Total School Budget': series_per_school_budget,
    'Per Student Budget': series_per_school_capita,
    'Average Math Score': series_per_school_math,
    'Average Reading Score': series_per_school_reading,
    '% Passing Math': series_per_school_passing_math,
    '% Passing Reading': series_per_school_passing_reading,
    '% Overall Passing': series_overall_passing_rate
}

# Creating and formatting a dataframe from the dictionary
df_per_school_summary = pd.DataFrame(dict_school_summary)
df_per_school_summary["Total School Budget"] = df_per_school_summary["Total School Budget"].map("${:,.2f}".format)
df_per_school_summary["Per Student Budget"] = df_per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
df_per_school_summary
```

## Scores by School Spending

```python
# Establish the bins 
list_spending_bins = [0, 585, 630, 645, 680]
list_spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
df_school_spending = df_per_school_summary.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
df_school_spending["Per Student Budget"] = df_school_spending["Per Student Budget"].astype(str).str.replace("$", "", regex = False).astype(float)

# cut spending based on the bins
df_school_spending["Spending Ranges (Per Student)"] = pd.cut(df_school_spending["Per Student Budget"], bins = list_spending_bins, labels = list_spending_labels)
df_school_spending["Per Student Budget"] = df_school_spending["Per Student Budget"].map("${:,.2f}".format)

df_school_spending
```

```python
#  Calculate averages for the desired columns. 
series_spending_math_scores = df_school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
series_spending_reading_scores = df_school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
series_spending_passing_math = df_school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
series_spending_passing_reading = df_school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
series_spending_overall_passing = df_school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
# Create a dictionary of the data
dict_shool_spending = {
    "Average Math Score": series_spending_math_scores,
    "Average Reading Score": series_spending_reading_scores,
    "% Passing Math": series_spending_passing_math,
    "% Passing Reading": series_spending_passing_reading,
    "% Overall Passing": series_spending_overall_passing
}

# Create the dataframe
df_spending_summary = pd.DataFrame(dict_shool_spending)

# Display results
df_spending_summary
```
